```python
def compute_cost(x,y,w,b):

    m = x.shape[0]

    cost = 0.0 

    for i in range(m):

        f_wb = w * x[i] + b 

        cost += (f_wb - y[i])**2

    total_cost = 1/(2*m) * cost 

    return total_cost

def compute_gradient(x,y,w,b):

    m = x.shape[0]

    dj_dw = 0
    dj_db = 0 

    for i in range(m):

        f_wb =  w * x[i] + b 

        dj_dw_i = (f_wb- y[i]) * x[i]

        dj_db_i = (f_wb- y[i])

        dj_dw += dj_dw_i
        dj_db += dj_db_i

    dj_dw = dj_dw/m
    dj_db = dj_db/m 


    return dj_dw,dj_db


def gradient_descent(x,y,w_in,b_in,num_iters,alpha,compute_cost,compute_gradient):


    J_history = []
    p_history = []

    w = w_in 
    b = b_in

    for i in range(num_iters):

        dj_dw, dj_db = compute_gradient(x,y,w,b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db 

        if num_iters < 10000:
            J_history.append(compute_cost(x,y,w,b))
            p_history.append([w,b])
        
            if i % 150 == 0: 
                print(f"Iteration{i}: Cost:{J_history[-1]}, w: {w}, b:{b}, dj_dw: {dj_dw}, dj_db:{dj_db}")

    return w,b,J_history,p_history
```

```python
w_init = 0
b_init = 0 

iterations = 1000

alpha_tmp = 1.0e-2
```

```python
w_final, b_final, J_history, p_history = gradient_descent(X_train, y_train, w_init, b_init, iterations, alpha_tmp, compute_cost, compute_gradient)

print(f"Gradient descent found these w:{w_final} and b:{b_final}")
```
